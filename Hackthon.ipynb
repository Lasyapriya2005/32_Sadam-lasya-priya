```python
required_columns = [
    'Hospital_Name',
    'District',
    'Location_Coordinates',
    'Emergency_Services',
    'Number_Doctor',
    'Total_Num_Beds',
    'Specialties',
    'Facilities'
]

df_clean = df[required_columns].copy()
df_clean.head()
```

```python
df_clean = df_clean.rename(columns={
    'Hospital_Name': 'hospital_name',
    'District': 'city',
    'Location_Coordinates': 'location_coordinates',
    'Emergency_Services': 'emergency_services',
    'Number_Doctor': 'number_doctor',
    'Total_Num_Beds': 'total_num_beds',
    'Specialties': 'specialties',
    'Facilities': 'facilities'
})

df_clean.columns.tolist()
```

```python
df_clean['location_coordinates'].head()
```

```python
df_clean[['latitude', 'longitude']] = (
    df_clean['location_coordinates']
    .astype(str)
    .str.split(',', expand=True)
)

df_clean['latitude'] = pd.to_numeric(df_clean['latitude'], errors='coerce')
df_clean['longitude'] = pd.to_numeric(df_clean['longitude'], errors='coerce')
```

```python
df_clean = df_clean.dropna(subset=['latitude', 'longitude'])
```

```python
import math

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c
```

```python
print("Enter accident / emergency location:")

user_lat = float(input("Enter latitude: "))
user_lon = float(input("Enter longitude: "))

incident_type = input("Enter incident type (General Emergency): ")

print("\nInput received:")
print("Latitude:", user_lat)
print("Longitude:", user_lon)
print("Incident:", incident_type)
```

```python
df_clean['distance_km'] = df_clean.apply(
    lambda row: haversine_distance(
        user_lat, user_lon,
        row['latitude'], row['longitude']
    ),
    axis=1
)

df_clean[['hospital_name', 'city', 'distance_km']].head()
```

```python
emergency_keywords = [
    'icu',
    'blood bank',
    'emergency ward',
    'trauma care',
    'emergency'
]
```

```python
df_clean['emergency_services_flag'] = df_clean['emergency_services'] \
    .astype(str) \
    .str.lower() \
    .apply(lambda x: any(keyword in x for keyword in emergency_keywords))
```

```python
df_clean['total_num_beds'] = pd.to_numeric(
    df_clean['total_num_beds'], errors='coerce'
).fillna(0)
```

```python
general_emergency_hospitals = df_clean[
    (df_clean['emergency_services_flag']) &
    (df_clean['total_num_beds'] > 0)
].copy()

print("Hospitals suitable for general emergency:", len(general_emergency_hospitals))
```

```python
general_emergency_hospitals['distance_score'] = (
    1 / (general_emergency_hospitals['distance_km'] + 1)
)

general_emergency_hospitals['bed_score'] = (
    general_emergency_hospitals['total_num_beds'] /
    general_emergency_hospitals['total_num_beds'].max()
)

general_emergency_hospitals['suitability_score'] = (
    0.7 * general_emergency_hospitals['distance_score'] +
    0.3 * general_emergency_hospitals['bed_score']
)

top_3 = general_emergency_hospitals.sort_values(
    by='suitability_score',
    ascending=False
).head(3)

top_3[
    ['hospital_name', 'city', 'distance_km',
     'total_num_beds', 'suitability_score']
]

```
